```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('advertising.csv')
```

```python
df.head()
```

```python
X=df[['TV','Radio','Newspaper']]
```

```python
y=df.Sales
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml = LinearRegression()
ml.fit(X,y)
```

```python
ml.intercept_
```

```python
ml.coef_
```

```python
ml.score(X,y)
```

```python
#statsmodels- a lib that gives the metrices in a stats model
import statsmodels.formula.api as smf
ml=smf.ols(formula='Sales~TV',data=df).fit()
ml.summary()
```

```python
import statsmodels.formula.api as smf
ml=smf.ols(formula='Sales~TV+Radio+Newspaper',data=df).fit()
ml.summary()
```
